```python
def separarColumnas(row):
  resultado = row['Resultado']
  evento = row['Evento']
  resultadoSeparado = resultado.split(":")

  if evento == "Home":
    return resultadoSeparado[0], resultadoSeparado[1]
  elif evento == "Away":
   return resultadoSeparado[1], resultadoSeparado[0]
```

```python

dfMessi[["Resultado Equipo", "Resultado Oponente"]] = dfMessi.apply(lambda x: separarColumnas(x), axis=1, result_type="expand")
```

üßπ Vamos a convertir en tipo n√∫merico el minuto del gol

```python
dfMessi[["Minuto"]]
```

```python
def convertirNumeros(x):
  if type(x) == str:
    string = x.split("+")
    string = [int(s.strip("'")) for s in string]
    return sum(string)
```

```python
dfMessi["Minuto Numerico"] = dfMessi["Minuto"].apply(lambda x: convertirNumeros(x))
```

üîé ¬øMete m√°s goles en primer o segundo tiempo?


```python
dfMessi.head()
```

```python
def enQueTiempo(row):
  if type(row) == str:
    string = row.split("+")
    string = [int(s.replace("'", "")) for s in string]
    if len(string) == 2:
      return "Primer tiempo extra" if string[0] == 45 else "Segundo tiempo extra"
    else:
      if string[0] <= 45:
        return "Primer tiempo"
      else:
        return "Segundo tiempo"
    
```

```python
dfMessi["Momento_del_gol"].unique()
```

```python
dfMessi.Minuto.unique()
```

```python
dfMessi['Momento_del_gol'] = dfMessi['Minuto'].apply(lambda x: enQueTiempo(x))
```

```python
golesPrimerTiempo= dfMessi[dfMessi.Momento_del_gol.isin(['Primer tiempo'])].count()[0] 
golesSegundoTiempo= dfMessi[dfMessi.Momento_del_gol.isin(['Segundo tiempo'])].count()[0]
golesPrimerTiempoExtra=dfMessi[dfMessi.Momento_del_gol.isin(["Primer tiempo extra"])].count()[0]
golesSegundoTiempoExtra=dfMessi[dfMessi.Momento_del_gol.isin(["Segundo tiempo extra"])].count()[0]

print(golesPrimerTiempo)
print(golesSegundoTiempo)
print(golesPrimerTiempoExtra)
print(golesSegundoTiempoExtra)
```

**üîé ¬øCuantos goles hizo Messi por partido? (Debemos sacar el promedio)**

> En est√° parte debemos agrupar los registros de goles por partido. Pueden pasar que dos goles o m√°s se realicen en un mismo partido


```python
golesPorPartido= dfMessi.groupby(pd.Grouper(key="Fecha")).count().mean()
golesPorPartidoPromedio = golesPorPartido.mean()
print(f"El promedio de goles por partido es de {golesPorPartidoPromedio}")
```
